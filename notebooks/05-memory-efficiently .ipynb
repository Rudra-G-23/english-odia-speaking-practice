```python
def selected_random_line(path, k):
    """
    Reservoir Sampling method used for memory efficiently.
        
    :param path: path of the file.
    :param k: num_line
    """
    selected_lines = []

    with open(path, 'r') as f:
        
        for i, line in enumerate(f):
            if i <  1:
                selected_lines.append(line)
            
            elif random.random() < k / (i +1):
                replace_idx = random.randrange(k)
                selected_lines[replace_idx] = line

    return selected_lines
```

```python
list_of_words = selected_random_line(path, k=1)
```

```python
def clean_list(list_of_words):
    
    result_list = []
    
    for item in list_of_words:
        words = item.split(',')
        
        for word in words:
            clean_word = word.strip().replace('\n', '')
            
            if clean_word:
                result_list.append(clean_word)

    return result_list
```

```python
clean_list(list_of_words)
```

```python
def question_generator(list_of_words):
    three_words = set(random.sample(list_of_words, 3))
    question_word = random.choice(list(three_words))
    return question_word, three_words
```

```python
list_of_words = clean_list(list_of_words)
```

```python
import random

def dict_question_generator(word_dict):
    # Select 3 random words from the dictionary keys
    choices = random.sample(list(word_dict.keys()), 3)
    # Pick one of them to be the question
    question_word = random.choice(choices)
    
    print(f"Define this word: {question_word}")
    print("Options:")
    for i, word in enumerate(choices):
        print(f"{i}: {word_dict[word]}")
    return question_word

# Usage
my_dict = {"Apple": "A fruit", "Dog": "An animal", "Desk": "Furniture"}
dict_question_generator(my_dict)
```
