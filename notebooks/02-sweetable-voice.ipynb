```python
import pyttsx3

engine = pyttsx3.init()
voices = engine.getProperty('voices')

for i, voice in enumerate(voices):
    print(i, voice.name, voice.id)
```

```python
engine = pyttsx3.init()

# Set sweet voice properties
engine.setProperty('rate', 145)      # slow & clear
engine.setProperty('volume', 0.9)    # soft but audible

voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)  # try female voice

engine.say("Hello! Welcome to learning English with Python.")
engine.runAndWait()
```
