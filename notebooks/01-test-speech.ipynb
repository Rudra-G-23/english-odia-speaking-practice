```python
import pyttsx3


engine = pyttsx3.init()


new_rate = 140
engine.setProperty('rate', new_rate)


new_volume = 0.75
engine.setProperty('volume', new_volume)
voices = engine.getProperty('voices')

text = "Excuse me, I would be grateful if you could listen for a moment. Thank you for your time."


engine.say(text)

engine.runAndWait()
```
